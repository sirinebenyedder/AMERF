```python
from google.colab import files
import requests
from PIL import Image
import matplotlib.pyplot as plt
```

```python
uploaded = files.upload()

# Afficher l'image avant l'envoi
image_path = "000002.jpg"
img = Image.open(image_path)

plt.figure(figsize=(8, 6))
plt.imshow(img)
plt.axis('off')
plt.title("Image à envoyer")
plt.show()

# Envoyer la requête à l'API
url = "https://nonantagonistic-ileana-nonradiating.ngrok-free.dev/predict"

with open(image_path, "rb") as f:
    files = {"image": f}
    headers = {"X-User-ID": "Sara"}
    r = requests.post(url, files=files, headers=headers)

# Récupérer et traiter la réponse
response = r.json()

# Afficher toutes les prédictions
if 'predictions' in response:
    preds = response['predictions']
    if isinstance(preds, list):
        print(", ".join(preds))  # Affiche: train, car, person, etc.
    else:
        print(preds)  # Pour un format différent
else:
    print("Aucune prédiction trouvée dans la réponse")
```
